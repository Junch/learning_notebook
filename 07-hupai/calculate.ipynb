```python
import pandas as pd
```

```python
df = pd.read_csv('hupai.txt', delim_whitespace=True, parse_dates=['month'])
df.month = df.month.dt.to_period('M')
df.head()
```

```python
import pathlib
p = pathlib.Path('.')
files = list(p.glob('*-*.txt'))

frames = []
months = []
for f in files:
    df = pd.read_csv(f.name, delim_whitespace=True, header=None, names=['time', 'price'])
    frames.append(df)
    months.append(pd.Period(f.stem, 'M'))

total = pd.concat(frames, keys=months)
total.shape
```

```python
total.reset_index(inplace=True)
total.rename(columns={'level_0':'month', 'level_1':'second'}, inplace=True)
table = pd.pivot_table(total, values='price', index=['month'], columns=['second'])
table[[x for x in range(40, 60)]]
```

```python
total.dtypes
```

```python
for i in range(0, 60):
    table[i] = table[60] - table[i]
table[[x for x in range(30, 50)]]
```

```python
arr = []
for i in range(0, 60):
    arr.append(table[i].value_counts())
    
ret = pd.concat(arr, axis=1)
ret.fillna(0, inplace=True)
for i in range(0, 60):
    ret[i]=ret[i].astype('int')

ret[[x for x in range(30,51)]]
```
