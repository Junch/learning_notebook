```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('account.csv', parse_dates=['reportdate'])
df.reportdate = df.reportdate.dt.to_period("Q")
```

```python
# https://www.youtube.com/watch?v=AFnUhiRXXWk
start = '2000Q1'
end = '2018Q1'
symbol = 'sh600585'
df2 = df.loc[df.symbol==symbol]
df2.set_index('reportdate', inplace=True)
df3 = df2[(df2.index.quarter==4) & (df2.index >= start) & (df2.index < end)]
plt.rcParams['figure.figsize'] = (6, 4)
df3[['netprofit']].plot.bar()
```

```python
start = pd.Period('2017Q3')
dfsmall = df[df.reportdate >= start]
dfsmall.set_index('symbol', inplace=True)
dfsmall.head()
```
